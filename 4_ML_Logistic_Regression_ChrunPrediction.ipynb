```python
import numpy as np
import pandas as pd
dataset = pd.read_csv("/content/ChurnPrediction.csv")
pd.set_option('display.max_columns', None) # to make sure you can see all the columns in output window
dataset.head()
```

```python
print(dataset.shape)
dataset.groupby('PastEmployee').size()
```

```python
print(dataset.info())
```

```python
print(dataset.describe())
```

```python
## Converting categorical features into numeric
def converter(column):
  if column=="Y" or column=="y":
    return 1
  else:
    return 0  
```

```python
dataset['OverTime']=dataset['OverTime'].apply(converter)
dataset['PastEmployee']=dataset['PastEmployee'].apply(converter)
dataset['Gender']=dataset['Gender'].map({'Female' :1, 'Male':0 })
print(dataset.head())
print(dataset.info())


```

```python
## Conveting few features using pd.get_dummies
catagorical_feature = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
final_data = pd.get_dummies(dataset, columns = catagorical_feature)
final_data.head()
```

```python
print(final_data.info())
```

```python
X = final_data.drop('PastEmployee', axis=1)
Y = final_data['PastEmployee']
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
# Normalizing numerical features so that each feature has mean 0 and variance 1
feature_scaler = StandardScaler()
X_scaled = feature_scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from imblearn.over_sampling import SMOTE  #Synthetic Minority Oversampling Technique
# imblearn library can be installed using pip install imblearn
from imblearn.pipeline import Pipeline #This pipeline is very similar to the sklearn 
#one with the addition of allowing samplers.
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
#import warnings.filterwarnings("ignore", category=FutureWarning)
# Implementing Logistic Regression
#Tuning eta0, max_iter, and alpha parameters and implementing cross-validation using Grid Search
#Constant by which the updates are multiplied.
model = Pipeline([
       ('balancing', SMOTE(random_state = 101)),
        ('classification', SGDClassifier(loss = 'log', penalty = 'elasticnet', random_state = 1))
   ])
#eta0double, default=1 - The initial learning rate for the constant, invscaling or adaptive schedules. T
#he default value is 0.0 as eta0 is not used by the default schedule optimal.
#alphafloat, default=0.0001
 #,
grid_param = {'classification__eta0': [.001,.01,.1,1,10,100], 'classification__max_iter' : [100,500,1000], 'classification__alpha': [.001, .01,.1, 1,10,100], 
              'classification__l1_ratio': [0,0.3,0.5,0.7,1]}

gd_sr = GridSearchCV(estimator=model, param_grid=grid_param, scoring='accuracy', cv=5)
```

```python
gd_sr.fit(X_scaled, Y)

best_parameters = gd_sr.best_params_
print("Best parameters: ", best_parameters)

best_result = gd_sr.best_score_ # Mean cross-validated score of the best_estimator
print("Best result: ", best_result)
```

```python
## Direct way of oversampling if you do not use pipeine

from collections import Counter
oversample = SMOTE(sampling_strategy='minority')

 
# fit and apply the transform
X_over, y_over = oversample.fit_resample(X_scaled, Y)
print(Counter(Y))
#X_over.shape
print(Counter(y_over))
print(X.shape)
X_over.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_over,y_over,test_size=0.1,random_state=3)
```

```python
model = SGDClassifier(loss = 'log', penalty = 'elasticnet', random_state = 1, alpha= 0.01, eta0= 0.001, l1_ratio=0.5, max_iter= 100)
```

```python
model.fit(X_train, y_train)
pred_lr=model.predict(X_test)
pred_lr
```

```python
from sklearn import metrics as mt
print(mt.accuracy_score(y_test, pred_lr))
print( mt.confusion_matrix(y_test, pred_lr))
```
